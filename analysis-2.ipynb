# Country summary analysis

```python
import pandas as pd

data = pd.read_excel('Ukraine_Support_Tracker_Release_13.xlsx', sheet_name='Country Summary ($)')

data.to_csv('country-summary-usd.csv', index=False)
```

```python
summary = pd.read_csv('country-summary-usd.csv', index_col=False)
```

```python
summary.head()
```

```python
columns = ['Country', 'Total bilateral commitments']
country_total = summary[columns]
country_total.sort_values(by='Total bilateral commitments', ascending=False)
country_total.columns = ['Country', 'Total bilateral commitments ($ billion)']
country_total.head()
# Data viz 1
```

```python
columns = ['Country', 'Financial commitments', 'Humanitarian commitments', 'Military commitments', 'Total bilateral commitments']
country_dist = summary[columns]
country_dist.columns = ['Country', 'Financial commitments ($ billion)', 'Humanitarian commitments ($ billion)', 'Military commitments ($ billion)', 'Total bilateral commitments ($ billion)']
country_dist.head()
# Data viz 2
```

```python
usa_dist = country_dist[country_dist['Country'] == 'United States']
usa_dist.columns = ['Country', 'Financial commitments ($ billion)', 'Humanitarian commitments ($ billion)', 'Military commitments ($ billion)', 'Total bilateral commitments ($ billion)']
usa_dist.head()
# Data viz 3
```

```python
columns = ['Country', 'Total bilateral commitments', 'GDP (2021)', 'Total bilateral commitments.1']
gdp_prop = summary[columns]
gdp_prop.columns = ['Country', 'Total bilateral commitments', 'GDP (2021)', 'Total bilateral commitments (% of GDP)']
gdp_prop.head()
# Data viz 4
```

```python
# Exporting to CSVs for use in Tableau
country_total.to_csv('country_total.csv', index=False)
country_dist.to_csv('country_dist.csv', index=False)
usa_dist.to_csv('usa_dist.csv', index=False)
gdp_prop.to_csv('gdp_prop.csv', index=False)
```
